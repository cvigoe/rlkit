```python
n= 30
T = 50
sparse_vecs = set()
NUM_TRIALS = 100
GAMMA = 0.99
sigma=.00001

bel = Belief(n,sigma=sigma)

returns = []
recoveries = np.zeros((NUM_TRIALS,2*T))

for i in range(n):
    sparse_vec = np.zeros(n)
    sparse_vec[i] = 1
    sparse_vecs.add(tuple(sparse_vec))

sparse_vecs = np.asarray(list(sparse_vecs))

for trial in range(NUM_TRIALS):
    print(trial)
    bel = Belief(n,sigma=sigma)    
    beta = bel.sample()
    prev = np.ones(n)
    ep_return = 0
    for i in beta:
        if i == 0:
            print('.', end="")
        else:
            print('*', end="")
            
    for t in range(T):
        x1, x2 = split(prev)

        y1 = (x1/np.sum(x1)) @ beta + (np.random.normal()*sigma)
        bel.filt(x1/np.sum(x1),y1)
        reward1 = bel.true_reward(beta)
        
        recovery1 = int(np.all(bel.sparse_vecs[np.argmax(bel.prior)] == beta))
        recoveries[trial][2*t] = recovery1
        
        y2 = (x2/np.sum(x2)) @ beta + (np.random.normal()*sigma)
        bel.filt(x2/np.sum(x2),y2)
        reward2 = bel.true_reward(beta)

        recovery2 = int(np.all(bel.sparse_vecs[np.argmax(bel.prior)] == beta))
        recoveries[trial,(2*t) + 1] = recovery2
        
        print('\r', end='')
        for i in x1:
            if i == 0:
                print('.', end="")
            else:
                print(1, end="")
        print(recovery1, '\r')
        for i in x2:
            if i == 0:
                print('.', end="")
            else:
                print(1, end="")
        print(recovery2, '\r')

        if y1 > y2:
            prev = x1
        else:
            prev = x2
            
        ep_return = ep_return + GAMMA**(t*2) * reward1
        ep_return = ep_return + GAMMA**(1+t*2) * reward2
    returns.append(ep_return)
```

```python
def split(x):
    x1 = np.zeros(len(x))
    x2 = np.zeros(len(x))
    length = np.sum(np.asarray(x) > 0)
    counter = 0
    if length == 1:
        return x, x
    if length == 0:
        raise Exception('Invalid split input vector!')
    for i in range(len(x)):
        if counter < int(length/2):
            x1[i] = x[i]
        else:
            x2[i] = x[i]
        counter += x[i]
    return x1,x2
```

```python
class Belief():
    def __init__(self,n=1,sigma=.1,eta=2,num_representation_draws=100, lmbd=1, itr=100):
        self.X = []
        self.y = []
        self.n = n
        self.sigma = sigma
        self.eta = eta
        self.num_representation_draws = num_representation_draws
        self.lmbd = lmbd
        self.itr = itr

        sparse_vecs = set()

        for i in range(self.n):
            sparse_vec = np.zeros(self.n)
            sparse_vec[i] = 1
            sparse_vecs.add(tuple(sparse_vec))

        self.sparse_vecs = np.asarray(list(sparse_vecs))
        self.prior = np.ones(len(self.sparse_vecs))/len(self.sparse_vecs)
    
    def filt(self,x,y):
        self.X.append(x)
        self.y.append(y)
        prior = self.prior
        posterior = np.zeros(len(self.sparse_vecs))
        for j in range(len(posterior)):
            posterior[j] =  self.gauss_pdf( y - self.sparse_vecs[j].T @ x )*prior[j]
        posterior = posterior/np.sum(posterior)
        self.prior = posterior

    def gauss_pdf(self,x):
        return np.exp(-0.5*((x/self.sigma)**2))
    

    def true_reward(self,beta):
        expectation = 0
        for index, sparse_vec in enumerate(self.sparse_vecs):
            expectation += self.prior[index] * np.mean((sparse_vec - beta)**2)
        return -1*expectation

    def sample(self):
        return self.sparse_vecs[np.random.choice(np.arange(len(self.sparse_vecs)), p=self.prior )]          
    
    def representation(self):
        return self.prior
    
```
